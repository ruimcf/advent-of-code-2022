```python
file = open("input.txt", "r")
text = file.read()
games = text.split("\n")

ROCK = "A"
PAPER = "B"
SISSOR = "C"

HAND_TO_POINT = {
    ROCK: 1,
    PAPER: 2,
    SISSOR: 3
}

LOSE = "X"
DRAW = "Y"
WIN = "Z"

RESULT_TO_POINT = {
    LOSE: 0,
    DRAW: 3,
    WIN: 6
}

acc = 0

hand_results = []
game_results = []

for game in games:
    opponent = game[0]
    requiredResult = game[2]

    myHand = 0
    if opponent == ROCK:
        if requiredResult == LOSE:
            myHand = SISSOR
        elif requiredResult == DRAW:
            myHand = ROCK
        else:
            myHand = PAPER
    
    if opponent == PAPER:
        if requiredResult == LOSE:
            myHand = ROCK
        elif requiredResult == DRAW:
            myHand = PAPER
        else:
            myHand = SISSOR
    
    if opponent == SISSOR:
        if requiredResult == LOSE:
            myHand = PAPER
        elif requiredResult == DRAW:
            myHand = SISSOR
        else:
            myHand = ROCK
    
    hand_point = HAND_TO_POINT[myHand]
    result_point = RESULT_TO_POINT[requiredResult]

    hand_results.append(hand_point)
    game_results.append(result_point)

print(sum(hand_results) +sum(game_results))

    
```
